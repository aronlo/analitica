```r
# importing modules
# ... YOUR CODE FOR TASK 1 ...
# Read datasets/yearly_deaths_by_clinic.csv into yearly
library(readr)
library(ggplot2)
yearly <- read_csv("./datasets/yearly_deaths_by_clinic.csv")
```

```r
# Print out yearly
yearly
```

```r
# Calculate the proportion of deaths per number of births and store the result in the new column 
yearly$proportion <- yearly$deaths/yearly$births
yearly$proportion
```

```r
# Extract clinic 1 data into yearly1 and clinic 2 data into yearly2

yearly1 <- subset(yearly, yearly$clinic=='clinic 1')

yearly2 = subset(yearly, yearly$clinic=='clinic 2')

# Plot yearly proportion of deaths at the two clinics
# plot the data using ggplot
ggplot(data = yearly, aes(x = year, y =  deaths)) + geom_line(aes(color = yearly$clinic)) 
```

```r
data_types <- function(frame) {
  res <- lapply(frame, class)
  res_frame <- data.frame(unlist(res))
  barplot(table(res_frame), main="Data Types", col="steelblue", ylab="Number of Features")
}


data_types(monthly)
```
