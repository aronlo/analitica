```r
library(ggplot2)

movies <- read.csv(file="./datasets/MovieRatings.csv", stringsAsFactors = TRUE ,header=TRUE, sep=",")

colnames(movies)<-c("Film","Genre","CriticalRating","AudienceRating","BudgetMillions","Year")
head(movies)
```

```r
ggplot(data= movies , aes(x=CriticalRating,y=AudienceRating, colour=Genre, size=BudgetMillions)) + geom_point()
```

```r
m<-ggplot(data= movies , aes(x=CriticalRating,y=AudienceRating, colour=Genre, size=BudgetMillions)) + geom_point()
m + geom_line()
```

```r
m + geom_line() + geom_point()
```

```r
m + geom_line(size=1) + geom_point()
```

```r
m + geom_point(aes(size=CriticalRating))
```

```r
m + geom_point() + xlab("Millions")
```

```r
###################   histograms  ##############################
s <- ggplot(data=movies,aes(x=BudgetMillions))
s + geom_histogram(binwidth = 10)
```

```r
s + geom_histogram(binwidth = 10, aes(fill=Genre))
```

```r
s + geom_histogram(binwidth = 10, aes(fill=Genre), colour="Black")
```

```r
s + geom_density(aes(fill=Genre))
```

```r
s + geom_density(aes(fill=Genre), position="stack")
```

```r
################# smooth ######################################
m<-ggplot(data= movies , aes(x=CriticalRating,y=AudienceRating, colour=Genre, size=BudgetMillions)) 
m + geom_point() + geom_smooth()
```

```r
m + geom_point() + geom_smooth(fill=NA)
```
