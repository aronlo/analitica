```python
import pandas as pd
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_excel('./datasets/Dataset_filtrado.xlsx')
```

```python
df
```

```python
df.drop(['Purpose', 'Checking_account'], axis=1, inplace = True)
```

```python
df
```

```python
df['Sex'] = df['Sex'].map({'female' : 1, 'male': 2})
df['Housing'] = df['Housing'].map({'own' : 1, 'free': 2 , 'rent': 3})
df['Saving_accounts'] = df['Saving_accounts'].map({'little' : 1, 'moderate': 2 , 'quite rich': 3, 'rich': 4})
df['Risk'] = df['Risk'].map({'bad' : 1, 'good': 2 })
df['Tipo'] = df['Tipo'].map({'Express A' : 1, 'Express B': 2 })
```

```python
df
```

```python
df.dropna(inplace = True)
```

```python
df
```

```python
X = df.iloc[:,0:8]
y = df.iloc[:,8]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42,stratify=df['Tipo'])
```

```python
log_clf = LogisticRegression(solver="liblinear", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
#svm_clf = SVC(gamma="auto", random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf)], voting='hard') #, ('svc', svm_clf)],
```

```python
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
#from sklearn.svm import SVC

log_clf = LogisticRegression(solver="liblinear", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
#svm_clf = SVC(gamma="auto", random_state=42)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf)], voting='hard') #, ('svc', svm_clf)],
```

```python
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf,  voting_clf): #svm_clf,
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```
